```python
import os
import GEOparse
import pandas as pd

os.chdir(os.path.expanduser('~/projects/lungcell'))

gse = GEOparse.get_GEO(geo="GSE40419", destdir="data/raw", silent=False)

print("Download complete!")
print(f"Number of samples: {len(gse.gsms)}")
print(f"\nFirst sample ID: {list(gse.gsms.keys())[0]}")
```

```python
expr_dict = {}
for gsm_id, gsm in gse.gsms.items():
    if gsm.table is not None and len(gsm.table) > 0:
        expr_dict[gsm_id] = gsm.table.set_index('ID_REF')['VALUE']

df = pd.DataFrame(expr_dict)
print("Expression matrix shape:", df.shape)
print("First 3 genes:", df.index[:3].tolist())
print("First 3 samples:", df.columns[:3].tolist())
print(df.iloc[:3, :3])
```

```python
labels = {}
for gsm_id, gsm in gse.gsms.items():
    metadata = gsm.metadata
    if gsm_id == list(gse.gsms.keys())[0]:  # print metadata for the first sample
        print("Available metadata fields:")
        for key, val in metadata.items():
            print(f"  {key}: {val}")
    break
```

```python
# Check source labels across every sample
source_counts = {}
for gsm_id, gsm in gse.gsms.items():
    source = gsm.metadata.get('source_name_ch1', ['unknown'])[0]
    source_counts[source] = source_counts.get(source, 0) + 1

print("Sample types found:")
for source, count in source_counts.items():
    print(f"  '{source}': {count} samples")
```

```python
# See what supplementary files the series has
print("\nSeries supplementary files:")
for key, val in gse.metadata.items():
    if 'supplementary' in key.lower():
        print(f"  {key}: {val}")
```

```python
import os
import pandas as pd
import GEOparse

os.chdir(os.path.expanduser('~/projects/lungcell'))

df = pd.read_csv('data/raw/GSE40419_RPKM.txt', sep='\t', comment='!', index_col=0)

print("=== Experession matrix ===")
print(f"Genes:   {df.shape[0]:,}")
print(f"Samples: {df.shape[1]}")
print("\nFirst 3 rows, 3 cols:")
print(df.iloc[:3, :3])
```

```python
# ── Extract tumour / normal labels from GEOparse metadata ──────────
# gse is already loaded from before — if kernel restarted, rerun:
# gse = GEOparse.get_GEO(geo="GSE40419", destdir="data/raw/", silent=True)

labels = {}
for gsm_id, gsm in gse.gsms.items():
    source = gsm.metadata.get('source_name_ch1', ['unknown'])[0]
    if 'normal' in source.lower():
        labels[gsm_id] = 'normal'
    else:
        labels[gsm_id] = 'cancer'

label_series = pd.Series(labels, name='condition')

print("\n=== Sample labels ===")
print(label_series.value_counts())
print("\nFirst 5 labels:")
print(label_series.head())
```

```python
# Check what the expression column names look like vs GSM metadata titles
print("Expression column names (first 5):")
print(df.columns[:5].tolist())

print("\nGSM title → label mapping (first 5):")
for gsm_id, gsm in list(gse.gsms.items())[:5]:
    title = gsm.metadata.get('title', ['?'])[0]
    cond  = labels[gsm_id]
    print(f"  {gsm_id}  title='{title}'  →  {cond}")
```

```python
# Map title → condition
title_to_condition = {}
for gsm_id, gsm in gse.gsms.items():
    title = gsm.metadata.get('title', ['?'])[0]
    title_to_condition[title] = labels[gsm_id]

# Apply to expression columns
condition = df.columns.map(title_to_condition)
print("\nConditions aligned to columns:")
print(pd.Series(condition, index=df.columns).value_counts())
```

```python
import os
os.chdir(os.path.expanduser('~/projects/lungcell'))

# Save the expression matrix
df.to_csv('data/processed/expression_matrix.csv')

# Save the labels
condition_series = pd.Series(
    {col: title_to_condition[col] for col in df.columns if col in title_to_condition},
    name='condition'
)
condition_series.to_csv('data/processed/sample_labels.csv')

print("Saved:")
print("  data/processed/expression_matrix.csv")
print("  data/processed/sample_labels.csv")
print(f"\nReady for Phase 3:")
print(f"  {df.shape[0]:,} genes × {df.shape[1]} samples")
print(f"  87 lung cancer  |  77 healthy lung  |  RPKM normalised")
```
